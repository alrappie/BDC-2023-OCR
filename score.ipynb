```python
import pandas as pd
import numpy as np
```

```python
def perhitungan_acc(target,pred):
    point = 0
    if len(target)>len(pred):
        for i in target:
            if i in pred:
                point += 1
    else:
        for i in pred:
            if i in target:
                point += 1
    
    return point/max(len(target),len(pred))
```

```python
def perhitungan_acc2(target,pred):
    point = 0
    if len(target)>len(pred):
        for i in range(len(pred)):
            if pred[i] == target[i]:
                point += 1
    else:
        for i in range(len(target)):
            if pred[i] == target[i]:
                point += 1
    
    return point/max(len(target),len(pred))
```
